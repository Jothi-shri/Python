```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()


# Encode multiple categorical columns
for col in ["Grade", "Result"]:
    df[col] = label_encoder.fit_transform(df[col])

print(df)


# df[["Grade","Result"]] = label_encoder.fit_transform(df[["Grade","Result"]])
# print(df)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Keep only numeric columns
data1 = df.drop(["Name"], axis=1)


# Correlation matrix
correlation_matrix = data1.corr()

# Plot heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
X=df.drop(['Name',"Total","StudentID","Grade"],axis=1)
Y=df["Total"]
Y
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
X_train
```

```python
X_test
```

```python
Y_train
```

```python
Y_test
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, Y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
Y_test
```

```python
import joblib

joblib.dump(model, 'evebatch.pkl')
```
